## Zadanie 5 - Doświadczenie
Do zbioru danych należy dodać kolumnę, która będzie mówiła o poziomie doświadczenia danej osoby, w sposób
słowny, przyjmując:

- Junior (do 3 lat doświadczenia)
- Mid (4 do 8 lat)
- Senior (9 do 12 lat)
- Ekspert (od 13 lat)

```python
df['Doświadczenie lat.'].min()
```

```python
df.isna().sum()
```

```python
condition = [df['Doświadczenie lat.'] <= 3,
            (df['Doświadczenie lat.'] >= 4) & (df['Doświadczenie lat.'] <= 8),
            (df['Doświadczenie lat.'] >= 9) & (df['Doświadczenie lat.'] <= 12),
            df['Doświadczenie lat.'] >= 13]

experience = ['Junior', 'Mid', 'Senior', 'Ekspert']

df['Poz. dośw.'] = np.select(condition, experience, 'Nieznany')
df.head()
```

## Zadanie 6 - Rok dołączenia
Kolumna reprezentująca rok dołączenia do firmy ma puste wartości. Uzupełnij brakujące wartości używając
średniej dla osób o takim samym poziomie doświadczenia (kolumna z Zadania 3), w skali całej firmy.

```python
# avg_year_join = [int(df[df['Poz. dośw.'] == x]['Rok roz. pracy'].mean()) for x in df['Poz. dośw.'].unique()]
# all_avg_exp = {k:v for (k,v) in zip(df['Poz. dośw.'].unique(), avg_year_join)}

# Groupby - grupuje bazę danych na podstawie unikalnych danych w przekazanej kolumnie
# Transform('mean') - wylicza średnią dla zgrupowanych unikalnych danych
avg_year_join = df.groupby('Poz. dośw.')['Rok roz. pracy'].transform('mean')

# Wypełniamy dane średnimi a następnie zmieniamy float na int
df['Rok roz. pracy'] = df['Rok roz. pracy'].fillna(avg_year_join).astype(int)
df.head(5)
```

## Zadanie 7 - Data Scientists
Który oddział firmy ma najwięcej Data Scientistów ?

```python
data_scientists = df[df['Departament'] == 'Data Science']
oddzialy_ds = data_scientists.groupby('Miasto firmy').size()
max_oddzial = oddzialy_ds.idxmax() 
print(f"Oddział posiadający najwięcej Data Scientistów: {max_oddzial}")
```
